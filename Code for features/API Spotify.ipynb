```python
def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify", playlist_id)
    items = results['items']
    while results['next']:
        results = sp.next(results)
        items.extend(results['items'])
    return [i["track"]["name"] for i in items]
```

```python
def get_playlist_artists(playlist_id):
    results = sp.user_playlist_tracks("spotify", playlist_id)
    items = results['items']
    while results['next']:
        results = sp.next(results)
        items.extend(results['items'])
    return [i["track"]["artists"][0]["name"] for i in items]
```

```python
def get_features(uris):
    
    danceability = []
    energy = []
    loudness = []
    speechiness = []
    acousticness = []
    instrumentalness = []
    valence = []
    tempo = [] 
    
    for uri in uris:

        features = sp.audio_features(uri)[0]
        
        danceability.append(features["danceability"])
        energy.append(features["energy"])
        loudness.append(features["loudness"])
        speechiness.append(features["speechiness"])
        acousticness.append(features["acousticness"])
        instrumentalness.append(features["instrumentalness"])
        valence.append(features["valence"])
        tempo.append(features["tempo"])        

    return danceability,energy,loudness,speechiness,acousticness,instrumentalness,valence,tempo
```
